```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 

data_path = "../data/processed/cleaned_products.csv"
df=pd.read_csv(data_path)

print("Basic Statistics")
print(df.describe())

plt.hist(df["price"], bins=20, edgecolor="skyblue")
y = np.linspace(df["price"].min(), df["price"].max(), 100)
coefficients = np.polyfit(df["price"], df["price"].value_counts(), 2)
polynomial = np.poly1d(coefficients)
plt.plot(y, polynomial(y), label='Pol√≠gono de frecuencias')
plt.title("Distribucion de precios")
plt.xlabel("Precio")
plt.ylabel("Producto")
plt.axvline(x=df["price"].mean(), color='r', linestyle='dashed', linewidth=2, label='Precios')
plt.legend()
plt.show()


print("Productos mas caros:")
highestPrice=df.nlargest(5,"price")
print(highestPrice)
```
